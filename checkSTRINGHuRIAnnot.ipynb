# Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Files

```python
mappings_STRING = pd.read_csv('DB/idmapping_2024_06_11STRING.tsv',sep='\t', header=0)
annotations_file_path = pd.read_csv('DB/HuriTest/goa_human.gaf',sep='\t', header=None)
```

```python
mappings_dict = {}

for index, row in mappings_STRING.iterrows():
    from_value = row['From']
    entry_value = row['Entry']
    
    if from_value in mappings_dict:
        mappings_dict[from_value].add(entry_value)
    else:
        mappings_dict[from_value] = set([entry_value])
        
mappings_dict
```

```python
data_dict = {}

for index, row in annotations_file_path.iterrows():
    key = row[1]
    value = row[4]
    
    if key in data_dict:
        data_dict[key].add(value)
    else:
        data_dict[key] = set([value])

final_dict = {}

for key,value in mappings_dict.items():
    for val in value:
        if val in data_dict:
            if key in final_dict:
                final_dict[key].update(list(data_dict[val]))
            else:
                final_dict[key] = set(list(data_dict[val]))
```

```python
result_dict_STRING = {key: len(value) for key, value in final_dict.items()}
result_dict_STRING
```

```python
import statistics

values = list(result_dict_STRING.values())
median = statistics.median(values)

print(median)
```

```python
# Get the values from the result_dict
values = list(result_dict_STRING.values())

# Create a boxplot
plt.boxplot(values)

# Set labels and title
plt.xlabel('Keys')
plt.ylabel('Values')
plt.title('Boxplot of Values in STRING')

# Show the plot
plt.show()
```
